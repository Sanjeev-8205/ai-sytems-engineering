# Logits and Softmax

```python
import seaborn as sns
import pandas as pd
import numpy as np
```

## Exp - 1

```python
vocabulary = {
    "Paris": 8.2,
    "London": 5.4,
    "Berlin": 2.8,
    "Rome": 2.1,
    "Banana": -1.4
}

df = pd.DataFrame(list(vocabulary.items()), columns=["City", "Value"])
df
```

```python
m = max(abs(df["Value"].min()), abs(df["Value"].max()))
ax = sns.barplot(data=df, x='City', y='Value', hue='City')
ax.set_ylim(-m, m)
```
